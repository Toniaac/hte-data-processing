```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('water_permeability_17.csv')
df
```

```python
df['water flux'] = df['Water Permeance (LMH/bar)'] * df['Presseure (bar)']
df
```

```python
# plot the data
# plot the water flux on a secondary y-axis
plt.plot(df['Presseure (bar)'], df['Initial permeance'], 's-', label='Initial Permeance')
plt.plot(df['Presseure (bar)'], df['Water Permeance (LMH/bar)'], 'o-', label='Steady State Permeance (after 8-12 hours)')
plt.xlabel('Pressure (bar)', fontsize=14)
plt.ylabel('Water Permeance (LMH/bar)', fontsize=14)
plt.title('Water Permeance vs Pressure of 17wt% PSf Membrane', fontsize=16)
# label the data points
for i in range(len(df)):
    plt.text(df['Presseure (bar)'][i], df['Initial permeance'][i], df['Initial permeance'][i], fontsize=10)
    plt.text(df['Presseure (bar)'][i], df['Water Permeance (LMH/bar)'][i], df['Water Permeance (LMH/bar)'][i], fontsize=10)
plt.legend()
plt.show()
```

```python
# plot again with two y-axes
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.plot(df['Presseure (bar)'], df['Initial permeance'], 's-', label='Initial Permeance')
ax1.plot(df['Presseure (bar)'], df['Water Permeance (LMH/bar)'], 'o-', label='Steady State Permeance (after 8-12 hours)')
ax2.plot(df['Presseure (bar)'], df['water flux'], '^-', c = 'g',label='Water Flux (LMH)')
ax1.set_xlabel('Pressure (bar)', fontsize=14)
ax1.set_ylabel('Water Permeance (LMH/bar)', fontsize=14)
ax2.set_ylabel('Water Flux (LMH)', fontsize=14)
plt.title('Water Permeance vs Pressure of 17wt% PSf Membrane', fontsize=16)
# label the data points
for i in range(len(df)):
    ax1.text(df['Presseure (bar)'][i], df['Initial permeance'][i], df['Initial permeance'][i], fontsize=10)
    ax1.text(df['Presseure (bar)'][i], df['Water Permeance (LMH/bar)'][i], df['Water Permeance (LMH/bar)'][i], fontsize=10)
    ax2.text(df['Presseure (bar)'][i], df['water flux'][i], round(df['water flux'][i]), fontsize=10)
ax1.legend(loc = 'upper left', fontsize=8)
ax2.legend(loc = 'upper right', fontsize=8)
ax1.set_ylim(0, 850)
ax2.set_ylim(0, 800)
plt.show()
```
